```python
# pip install matplotlib (in terminal)

# import matplolib
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

# 1. Line plot

```python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]


# 1. create a line plot 
plt.plot(x, y, color='r', linewidth = 5, linestyle = '--', label = 'profit')

# 2. configure the plot 
plt.legend()
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('My first line plot')

# 3. save the plot 
plt.savefig('./exports/lineplot1.png')

# 4. display plot
plt.show()
```

# 2. Scatter Plot

```python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# create a DataFrame

df = pd.DataFrame( {'X': x, 'y': y})

# create a line plot
plt.scatter(df['X'], df['y'], marker = '^', s = 50, color = 'red', label = 'my marker') # s = size 

# configure the plot 
plt.legend()
plt.grid(True)
plt.xlabel('data_point')
plt.ylabel('price')
plt.title('My second line plot')

# display plot 
plt.show()
```

# 3. Bar Plot

# 4. Histogram

# 5. Pie Chart

# 3D Surface Plot

```python
import numpy as np

# generate data for the surface plot
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# create the 3 D Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(X, Y, Z, cmap = 'viridis')

# add labels and title
ax.set_xlabel('X_axis')
ax.set_ylabel('Y_axis')
ax.set_zlabel('Z_axis')
ax.set_title('3D Surface Plot')
```

# Subplots

```python
"""
Line Styles : '-', '--', ':', 'None', 'dashed', 'dashdot', 'dotted'
"""

# data
x = [1,2,3,4,5,6,7]
y= [10,23,45,67,89,2,14]
z = [5,13,35,27,69,20,14]

# subplot (row_count, column_count, index of the axes )
plt.subplot(2,2,1)
plt.plot(x,y, color='blue')

plt.subplot(2,2,2)
plt.plot(x,z, color='green')

plt.subplot(2,2,3)
plt.plot(x,z, color='red')

plt.subplot(2,2,4)
plt.plot(x,z, color='black')

# show the plot 
plt.show()
```

```python
"""
Line Styles : '-', '--', ':', 'None', 'dashed', 'dashdot', 'dotted'
"""

# data
x = [1,2,3,4,5,6,7]
y= [10,23,45,67,89,2,14]
z = [5,13,35,27,69,20,14]

fig, ax = plt.subplots(2,2)
# subplot (row_count, column_count, index of the axes )

# row 0
ax[0,0].plot(x,y, color = 'blue')
ax[0,1].plot(x, z, color = 'green')


# row 1
ax[1, 0].plot(x,z, color = 'red')
ax[1,1].plot(x,z, color = 'black')


# show the plot 
plt.show()
```

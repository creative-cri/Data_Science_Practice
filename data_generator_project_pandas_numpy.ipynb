### Write a python program which generates a CSV file with following columns:
1. ID: values from 1 to 100
2. Product title: a random choice from ["Brot", "Kaese", "Honig"]

```python
# 1. create ID: values from 1 to 100 + 
ID = np.arange(1, 101)
ID
```

```python
# 3. create price: a random price between 1€ and 4€ + 
price = np.random.uniform(1, 4, size=100) 
print(price)
```

```python
# create the Data in a dictionary 

data = {
    'ID': ID,
    'Product Title':product_title,
    'Price': price
}

# create the DataFrame
df = pd.DataFrame(data)
df
```

### 3. Convert the scripts into functions-wise and optional to object oriented-wise

```python
# Function-wise approach

def generate_csv_with_numpy():
    # Generate data using Numpy
    id = np.arange(1, 101)
    product_titles = np.random.choice(["Brot", "Kaese", "Honig"], size=100)
    prices = np.round(np.random.uniform(1, 4, size=100), decimals=2)

    # Create DataFrame using Pandas
    df = pd.DataFrame({'ID': id, 'Product title': product_titles, 'Price': prices})

    # Calculate Brutto 7%
    df['Brutto 7%'] = (df['Price'] * 1.07).round(2)

     # Save to CSV file
    df.to_csv('products_with_brutto.csv', index=False)

generate_csv_with_numpy()
```

```python
# Object-oriented approach:


import numpy as np
import pandas as pd

class ProductGenerator:
    
   def __init__(self):
       pass
    
   def generate_data(self):
       self.ids = np.arange(1,101)
       self.product_titles = np.random.choice(["Brot", "Kaese", "Honig"], size=100)
       self.prices =np.round(np.random.uniform(1 ,4 ,size=100), decimals=2)

   def create_dataframe(self):
        self.df=pd.DataFrame({'ID':self.ids,'Product title':self.product_titles,'Price':self.prices})
        
   def calculate_brutto(self):
        self.df['Brutto 7%']=(self.df['Price']*1.07).round(2)
   
   def save_to_csv(self,filename):
        self.df.to_csv(filename,index=False)


product_obj=ProductGenerator()
product_obj.generate_data()
product_obj.create_dataframe()
product_obj.calculate_brutto()
product_obj.save_to_csv('products_with_brutto.csv')
```

# Option to do this without Pandas

```python
import numpy as np

# Generate product data
np.random.seed(0)

NUM_ROWS = 100
LIST_OF_PRODUCTS =['Brot','Kaese','Honig']
MIN_PRICE = 1
MAX_PRICE = 4


product_ids = np.arange(1,NUM_ROWS+1) 
product_titles = np.random.choice(LIST_OF_PRODUCTS, size= NUM_ROWS)
product_prices = np.random.uniform(MIN_PRICE,MAX_PRICE,size= NUM_ROWS)

# Convert data types and format Price Column
product_ids_str = product_ids.astype(str)
product_titles_str = product_titles.astype(str)
product_prices_formatted = np.around(product_prices, decimals= 2)

# Combine data into an arry
product_data = np.column_stack( (product_ids_str,product_titles_str,product_prices_formatted))

# save data to a CSV file
header = 'ID,Product title, Price'
np.savetxt('product_data3.csv',product_data,delimiter=',', header= header,comments='',fmt='%s')
```

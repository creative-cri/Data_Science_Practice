# Apply
df.apply(function)

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(
    {'number1': [10, 20],
    'number2':[100, 200]

    }
)

print(df)
print()

# create function
def increment(x):
    return x+50

# call the function
increment(10) # 10 + 50 = 60 

# use the function on the data frame
df2 = df.apply(increment)
df2

# Alternative: use a lambda function on the dataframe
df3 = df.apply(lambda x: x+50)
df3

# Using numpy Funcction on the dataframe
df4 = df.apply(np.sum, axis=0) # sum over the rows
print(df4)
print()

df5 = df.apply(np.sum, axis=1) # sum over the columns
print(df5)
```

# Group By

```python
participants = [
    (10, 'thomas', 40, 'Berlin', 77),
    (11, 'Ingo', 40, 'Berlin', 88),
    (12, 'Sara', 40, 'Berlin', 90),
    (13, 'Lena', 30, 'Frankfurt', 70),
    (15, 'Frank', 20, 'Stuttgart', 78),
    (16, 'Matthias', 20, 'Aachen', 77),
    (17, 'Ali', 40, 'Aachen', 90),
    (18, 'timo', 40, 'Stuttgart', 77),
    (19, 'Steffi', 40, 'Stuttgart', 77),
    
]

# create a DataFrame
df = pd.DataFrame(participants, columns = ['ID', 'name', 'age', 'city', 'score'])

# set the index
df.set_index('ID', inplace = True)
df
```

```python
# Create groups based on a column
groups = df.groupby('city')
#groups = df.groupby('age')
groups


# Loop over the groups
for group_name, rows in groups:
    print('Group Name: ', group_name)
    print('Group Rows ', )
    print(rows, "\n")
```

```python
# Get the count of rows of each group
print(groups.size())
print()

# Aggregate Functions
mean_values = groups.mean(numeric_only = True)
print(mean_values)
```

```python
# do aggregation functions on each column separately
results = groups.agg(
    {
        'age': 'mean',
        'score': 'sum'
    }
)

results = groups['age'].agg( ['sum', 'mean', 'max'] )
results
```

```python
results = groups[['age', 'score']].agg(['sum', 'mean', 'max'])
results
```
